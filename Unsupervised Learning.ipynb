## Part 1 -- Transformation

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# one example of Transformation is Rescaling of Data, there are many techniques.
# the following focuses on Standardization or rescaling the data so that each feature
# is centered at 0 (mean = 0) with standard dev = 0.
```

```python
# manual standardization

ary = np.array([1,2,3,4,5])
ary_standardized = (ary - ary.mean()) / ary.std()
ary_standardized
```

```python
# now trying standardization on Iris data set with ScikitLearn

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, random_state=0)
print(X_train.shape)
```

```python
# the iris data is not "Centered" and has a diff std dev for each component
print "mean : %s " % X_train.mean(axis=0)
print "standard dev: %s" % X_train.std(axis=0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
# As with Classification and Regression we can fit a model
# Scaler is unsupervised so only X is passed

scaler.fit(X_train)
```

```python
# Now the data can be rescaled

X_train_scaled = scaler.transform(X_train)
```

```python
print X_train_scaled.shape
```

```python

print("mean : %s " % X_train_scaled.mean(axis=0))
print("standard deviation : %s " % X_train_scaled.std(axis=0))
```

```python
# Principal Component Analysis
#reduces dimensionality of data by creating a linear projection
rnd = np.random.RandomState(5)
X_ = rnd.normal(size=(300,2))
X_blob = np.dot(X_, rnd.normal(size=(2,2))) + rnd.normal(size=2)
y = X_[:, 0] > 0
plt.scatter(X_blob[:, 0], X_blob[:, 1], c=y, linewidths=0, s=30)
plt.xlabel('feature 1')
plt.ylabel('feature 2')
```

```python
from sklearn.decomposition import PCA
pca = PCA()
```

```python
pca.fit(X_blob)
```

```python
X_pca = pca.transform(X_blob)

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, linewidths=0, s=30)
plt.xlabel('first principal component')
plt.ylabel('second principal component')
```

## Part 2 -- Clustering


```python
# Clustering is the task of gathering samples into groups of similar samples according
# to some predefined similarity or distance (dissimilarity) measure.

#Common applications for clustering algorithms.
#    * Compression for data reduction
#    * Summarizing data as a reprocessing step for recommender systems
#    * Similarly:
#        * grouping related web news (e.g. Google News) and web search results
#        * grouping related stock quotes for investment portfolio management
#        * building customer profiles for market analysis
#    * Building a code book of prototype samples for unsupervised feature extraction
```

```python
# Start with a 2-dim, synthetic dataset

from sklearn.datasets import make_blobs

X, y = make_blobs(random_state = 42)
X.shape
```

```python
plt.scatter(X[:, 0], X[:, 1])
```

```python
# In the scatter plot there are apparent groups of data points
# Using clustering, class labels can be "discovered" for each group

# KMeans clustering based off distance from points to a given point, uses 
# Euclidean Distance for calculations

from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
```

```python
labels = kmeans.fit_predict(X)
```

```python
labels
```

```python
all(y == labels)
```

```python
plt.scatter(X[:,0], X[:,1], c=labels)
```

```python
plt.scatter(X[:, 0], X[:,1], c=y)
```

```python
# to check accuracy of the clustering

from sklearn.metrics import confusion_matrix, accuracy_score

print 'Accuracy score:', accuracy_score(y, labels)
print confusion_matrix(y, labels)
```

```python
np.mean(y == labels)
```

```python
## Why is the accuracy 0.0 and not 1.0??

# The cluster IDs were arbitrary, therefore a different scoring metric must be used,
# such as adjusted_rank_score which is invariant to permutations of labels

from sklearn.metrics import adjusted_rand_score

adjusted_rand_score(y, labels)
```

```python
# One shortcoming of KMeans is that the number of clusters has to be set, which is often
# unknown. 
# For example, if the KMeans was attempted with n_clusters = 2:

kmeans = KMeans(n_clusters=2, random_state=42)
labels = kmeans.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels)
```

```python
#The Elbow Method:
#     The elbow method is an approach of finding the optimal number of clusters

distortions = []
for i in range(1,11):
    km = KMeans(n_clusters=i,
               random_state=0)
    km.fit(X)
    distortions.append(km.inertia_)
plt.plot(range(1,11), distortions, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Distortion')
plt.show()

print "Pick the result that looks like the 'pit of an elbow' (3)"
```
