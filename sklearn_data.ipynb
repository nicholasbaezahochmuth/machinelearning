```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris.keys()
```

```python
n_samples, n_features = iris.data.shape
print('Number of samples:', n_samples)
print('Number of features:', n_features)
## The sepal length, sepal width, petal length, and petal width for the first sample
print(iris.data[0])
```

```python
print(iris.data.shape)
print(iris.target.shape)
```

```python
# List of the species of each data point

print(iris.target)
```

```python
import numpy as np
# using bincount one can count how many data points
# are in each species
#    class 0: Iris-Setosa
#    class 1: Iris-Versicolor
#    class 2: Iris-Virginica

np.bincount(iris.target) 
```

```python
# names for the species are stored in target_names

print(iris.target_names)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

# The data is four-dimensional but using MPL one can analyze one or two dimensions 
# at a time
```

```python
# Create histogram of Petal Width for each species of Iris

x_index = 3 # setting X_Index to 3 chooses "Petal Width" as the data
colors = ['blue', 'red', 'green']

for label, color in zip(range(len(iris.target_names)), colors):
    plt.hist(iris.data[iris.target==label, x_index],
            label=iris.target_names[label],
            color=color)
plt.xlabel(iris.feature_names[x_index])
plt.legend(loc='upper right')
plt.show()
```

```python
# Scatter Sepal Length vs Petal Width for each species of Iris

x_index = 3 # 3 selects "Petal Width" for x-axis
y_index = 0 # 0 selects "sepal length" for y-axis

colors = ['blue', 'red', 'green']

for label, color in zip(range(len(iris.target_names)), colors):
    plt.scatter(iris.data[iris.target==label, x_index],
               iris.data[iris.target==label, y_index],
               label=iris.target_names[label],
               c=color)
plt.xlabel(iris.feature_names[x_index])
plt.ylabel(iris.feature_names[y_index])
plt.legend(loc='upper left')
plt.show()
```

```python
# Scatter plot matrices
# Using Pandas, one can create scatter plots for each possible comparison

import pandas as pd

iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
pd.tools.plotting.scatter_matrix(iris_df, figsize=(8,8))
```

```python
# Loading Digits Data

from sklearn.datasets import load_digits
digits=load_digits()
```

```python
digits.keys()
```

```python
n_samples, n_features = digits.data.shape
print((n_samples, n_features))
```

```python
print(digits.data[0])
print(digits.target)
```

```python
print(digits.data.shape)
print(digits.images.shape)
```

```python
# data and images related by a reshaping

import numpy as np
print(np.all(digits.images.reshape((1797,64)) == digits.data))
```

```python
# visualization is more intensive

# to set up the figure:
fig = plt.figure(figsize=(6,6))
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)

# plot the digits, each image is 8x8 pixels
for i in range(64):
    ax = fig.add_subplot(8,8, i+1, xticks=[], yticks=[])
    ax.imshow(digits.images[i], cmap=plt.cm.binary, interpolation='nearest')
    
    #label the image with target val
    ax.text(0, 7, str(digits.target[i]))
```
